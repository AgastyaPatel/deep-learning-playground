# Hot and Cold Learning

```python
# Simple form of gradient descent
weight = 0.5
inp = 0.5
goal_prediction = 0.8

step_amount = 0.01

for iteration in range(1101):
    prediction = inp * weight
    error = (prediction - goal_prediction) ** 2

    print("Error: " + str(error) + " Prediction: " + str(prediction))

    up_prediction = inp * (weight + step_amount)
    up_error = (goal_prediction - up_prediction) ** 2
    
    down_prediction = inp * (weight - step_amount)
    down_error = (goal_prediction - down_prediction) ** 2

    if (down_error < up_error):
        weight = weight - step_amount
    if (down_error > up_error):
        weight = weight + step_amount
```

### Caclculating both direction and amount
This addresses the problem of direction, amount/scaling and stopping.

```python
# Simple form of gradient descent
weight = 0.5
goal_pred = 0.8
inp = 0.5

for iteration in range(20):
    pred = inp * weight
    error = (pred - goal_pred) ** 2
                           # Pure Error       # Scaling, negative reversal and stoping
    direction_and_amount = (pred - goal_pred) * inp
    weight = weight - direction_and_amount
    print("Error: " + str(error) + " Prediction: " + str(pred))
```

### One Iteration of Gradient Descent

```python
weight = 0.1
alpha = 0.01    # Learning Rate

def neural_network(inp, weight):
    pred = inp * weight
    return pred

number_of_toes = [8.5]
win_or_lose_binary = [1]

inp = number_of_toes[0]
goal_pred = win_or_lose_binary[0]

pred = neural_network(inp, weight)
error = (pred - goal_pred) ** 2

delta = pred - goal_pred    # Raw amount if the node is high or low
weight_delta = delta * inp

print(pred, error,weight_delta)
```

```python
weight -= weight_delta * alpha
print(weight)
```

#### Multiple Iterations of Gradient Descent

```python
weight, goal_pred, inp = (0.5, 0.8, 0.5)

for iteration in range(200):
    pred = neural_network(inp, weight)
    error = (pred - goal_pred) ** 2 # Error
    delta = pred - goal_pred    # direction/Pure error
    weight_delta = delta * inp  # Scaling
    weight -= weight_delta * alpha
    print("Error: " + str(error) + " Prediction: " + str(pred))
```

```python
weight, goal_pred, inp = (0.0, 0.8, 1.1)

for iteration in range(4):
    print('------\nWeight: ' + str(weight))
    pred = inp * weight
    error = (pred - goal_pred) ** 2
    delta = pred - goal_pred
    weight_delta = delta * inp
    weight -= weight_delta
    print('Error: ' + str(error) + ' Prediction: ' + str(pred))
    print('Delta: ' + str(delta) + ' Weight Delta: ' + str(weight_delta))
```
